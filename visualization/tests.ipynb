```python
from pathlib import Path
import sys
import numpy as np
import matplotlib.pyplot as plt

cwd = Path.cwd().resolve()
project_root = cwd if (cwd / "data_generation" / "generate.py").exists() else cwd.parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from visualize import *
from data_generation.generate import *


shape_generator = ShapeGenerator()
```

```python
shape_start = shape_generator.build_shape(
    shape_name="star",
    num_points=180,
    center=(0.0, 0.0),
    scale=1.0,
    angle=np.deg2rad(4.0),
    n_tips=6,
    inner_radius=0.45,
).to_arrays()

shape_target = shape_generator.build_shape(
    shape_name="circle",
    num_points=180,
    center=(0.0, 0.0),
    scale=0.58,
    angle=0.0,
).to_arrays()

v_start, l_start = shape_start
v_target, l_target = shape_target
n_source = len(v_start)
n_target = len(v_target)

nearest_idx = nearest_point_matching(v_start, v_target)

matchings = nearest_idx.astype(int)
print(matchings[:10])
```

```python
visualize_matching(
    shape_start,
    shape_target,
    matching_mode="nearest",
    max_lines=180,
    point_size=20,
    line_width=0.42,
    line_alpha=0.33,
)
```

```python

visualize_matching(
    shape_start,
    shape_target,
    matching_mode="nearest",
    max_lines=180,
    show_arrows=True,
    arrow_percentage=30 / 180,
    point_size=18,
    line_width=0.42,
    line_alpha=0.32,
)
```

## Extra Examples

```python
# Provided matching + sparse arrows at 20% (divisor-based arrow count).
visualize_matching(
    shape_start,
    shape_target,
    matching=matchings,
    mode="both",
    matching_mode="provided",
    max_lines=180,
    show_arrows=True,
    arrow_percentage=0.20,
    line_width=0.42,
    line_alpha=0.40,
    point_size=8,
)
```

```python
examples = [
    {
        "start": dict(shape_name="star", num_points=180, n_tips=6, inner_radius=0.45, scale=1.0, angle=np.deg2rad(4.0)),
        "target": dict(shape_name="circle", num_points=180, scale=0.56, center=(0.0, 0.0)),
        "arrow_pct": 30 / 180,
    },
    {
        "start": dict(shape_name="hexagon", num_points=180, scale=1.0, angle=np.deg2rad(8.0)),
        "target": dict(shape_name="circle", num_points=180, scale=0.64, center=(0.10, -0.04)),
        "arrow_pct": 24 / 180,
    },
    {
        "start": dict(shape_name="triangle", num_points=180, scale=1.0, angle=np.deg2rad(-5.0)),
        "target": dict(shape_name="star", num_points=180, n_tips=6, inner_radius=0.38, scale=0.72, angle=np.deg2rad(10.0)),
        "arrow_pct": 36 / 180,
    },
]

fig, axes = plt.subplots(1, len(examples), figsize=(15, 5))

for ax, ex in zip(axes, examples):
    start = shape_generator.build_shape(**ex["start"]).to_arrays()
    target = shape_generator.build_shape(**ex["target"]).to_arrays()

    visualize_matching(
        start,
        target,
        mode="both",
        matching_mode="nearest",
        max_lines=180,
        show_arrows=True,
        arrow_percentage=ex["arrow_pct"],
        line_width=0.40,
        line_alpha=0.34,
        point_size=11,
        ax=ax,
    )

plt.tight_layout()
plt.show()
```
